### Python Data Types

#### Built-in Data Types
- In programming, data type is an important concept.

- Variables can store data of different types, and different types can do different things.

- Python has the following data types built-in by default, in these categories:

 |Data types | Categories                    |
|--------------|-------------------------------|
| Text Type     | 	str(string)                  |
| Numeric Types: | 	int, float, complex          |
| Sequence Types:| 	list, tuple, range           |
| Mapping Type:  | 	dict (dictionary)            |
| Set Types:	   | set, frozenset                |
| Boolean Type:  | bool   (boolean)              |
| Binary Types:  | 	bytes, bytearray, memoryview |
| None Type:     | 	NoneType                     |

#### Getting the Data Type
You can get the data type of any object by using the type() function:

```python

# Example: Print the Data Type of the variable x:

x = 5
print(type(x))
```

#### Setting the Data Type
In Python, the data type is set when you assign a value to a variable:

```python
# String

x = "Hello World"

# Display x:
print(x)

# Display the data type of x:
print(type(x))
```

```python
# Integer

x = 20
print(x)
print(type(x))
```

```python
# Floating

x = 20.5
print(x)
print(type(x))
```

```python
# Complex

x = 1j
print(x)
print(type(x))
```

```python
# List

x = ["apple", "banana", "cherry"]
print(x)
print(type(x))
```

```python
# Tuple

x = ('apple', 'banana', 'cherry')
print(x)
print(type(x))
```

```python
# Range

x =range(6)
print(x)
print(list(x))
print(type(x))
```

```python
# Dictionary

x = {"name":"John", "age":25}
print(x)
print(type(x))
```

```python
# Set

x = {"apple", "banana", "cherry"}
print(x)
print(type(x))
```

```python
# Frozenset

x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
```

```python
# Boolean
# True or False

x = True
print(x)
print(type(x))
```

```python
# Bytes

x = b"Hello"
print(x)
print(type(x))
```

```python
# Bytearray

x = bytearray(5)
print(x)
print(type(x))
```

```python
# Memory View

x = memoryview(bytes(5))
print(x)
print(type(x))
```

```python
# None Type

x = None
print(x)
print(type(x))
```

#### Setting the Specific Data Types

If you want to specify the data type, you can use the following constructor function:

```python
# String

x = str("Hello World")

# Display x:
print(x)

# Display the data type of x:
print(type(x))
```

```python
# Integer

x = int(20)
print(x)
print(type(x))
```

```python
# Floating

x = float(20.5)
print(x)
print(type(x))
```

```python
# Complex

x = complex(1j)
print(x)
print(type(x))
```

```python
# List

x = list(("apple", "banana", "cherry"))
print(x)
print(type(x))
```

```python
# Tuples

x = (("apple", "banana", "cherry"))
print(x)
print(type(x))
```

```python
# Range

x = range(3,10)
print(x)
print(list(x))
print(type(x))
```

```python
# Dictionary

x = dict(name="John", age=25)
print(x)
print(type(x))
```

```python
# Set

x = set(('apple', 'banana', 'cherry'))
print(x)
print(type(x))
```

```python
# Frozenset

x = frozenset(("apple", "banana", "cherry"))
print(x)
print(type(x))
```

```python
# Boolean

x = bool(5)
print(x)
print(type(x))
```

```python
# Bytes

x = bytes(5)
print(x)
print(type(x))
```

```python
# Bytearray

x = bytearray(5)
print(x)
print(type(x))
```

```python
# Memory View

x = memoryview(bytes(5))
print(x)
print(type(x))
```
