### Python Numbers

There are three numeric types in Python:

- int
- float
- complex

Variables of numeric types are created when you assign a value to them:

```python
# Example

x = 1    #int
y = 2.8  #float
z = 1j   # complex
```


To verify the type of any object in Python,use the type() function:

```python
print(type(x))
print(type(y))
print(type(z))
```

#### Integer

int or integer is a whole number,positive or negative without decimals of unlimited length.

```python
# Example

x = 1
y = 9856454521555123564
z = -8654231
```

```python
print(type(x))
print(type(y))
print(type(z))
```

#### Floating

float or floating pointer number is a number, positive or negative,containing one or more decimals.

```python
# Example

x = 9.10
y = 1.0
z = -45.59
```

```python
print(type(x))
print(type(y))
print(type(z))
```

Float can also be scientific numbers with an "e" to indicate the power of 10.

```python
# Example

x = 35e3
y = 15E5
z = -89.7e100
```

```python
print(type(x))
print(type(y))
print(type(z))
```

#### Complex

Complex numbers are written with a "j" as the imaginary part:

```python
# Example

x = 3+5j
y = 5j
z = -6j
```

```python
print(type(x))
print(type(y))
print(type(z))
```

#### Type Conversion

You can convert from one type to another with the int(),float() and complex() methods:

```python
# Convert from one type to another:

x = 1   #int
y = 2.5 #float
z = 1j  #complex
```

```python
print(type(x))
print(type(y))
print(type(z))
```

```python
# Convert from int to float
a = float(x)
print(a)
```

```python
print(type(a))
```

```python
# Convert from float to int:

b = int(y)
print(b)
print(type(b))
```

```python
# Convert from int to complex:
c = complex(x)
print(c)
print(type(c))
```

```python
print(a)
print(b)
print(c)
```

```python
print(type(a))
print(type(b))
print(type(c))
```

#### Note: You cannot convert complex number into another number type.

#### Random Number

Python does not have a random() function to make a random number, but Python has a built-in module called random that can be used to make random numbers:

### Python Casting

#### Specify a Variable Type

There may be times when you want to specify a type on to a variable. This can be done with casting. Python is an object-orientated language, and as such it uses classes to define data types, including its primitive types.

Casting in python is therefore done using constructor functions:

- int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)

- float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)

- str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals

```python
# Integers:

x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3
```

```python
print(x)
print(y)
print(z)
```

```python
# Floats:


w = float(1)     # w will be 1.0
x = float(2.5)   # x will be 2.0
y = float("4")   # y will be 4.0
z = float("5.2") # z will be 5.2
```

```python
print(w)
print(x)
print(y)
print(z)
```

```python
# Strings:

w = str('5.8') # w will be '5.8'
x = str("a1")  # x will be 'a1'
y = str(2)     # y will be '2'
z = str(4.0)   # z will be '4.0'
```

```python
print(w)
print(x)
print(y)
print(z)
```
